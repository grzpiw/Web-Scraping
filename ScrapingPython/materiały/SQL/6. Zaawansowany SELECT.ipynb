
``` SQL
SELECT id, brand FROM cars 
WHERE brand LIKE 'Fo___'
```


### GROUP BY

``` SQL
SELECT COUNT(*), fuel_type FROM cars
GROUP BY fuel_type
```

``` SQL
SELECT AVG(price), brand, COUNT(*) AS cnt FROM cars
GROUP BY brand
```

``` SQL
SELECT round(AVG(price), 2) as AVERAGE, brand FROM cars
GROUP BY brand
```

### CASE (WHEN THEN)

``` SQL
select price,
	  case
	  	when price < 20000 then 'Tani'
	    when price >= 20000 and price < 50000 then 'Średni'
	    else 'Drogi' end
from cars
```

``` SQL
select price,
	  case
	  	when price < (select max(price)/3 from cars) then 'Tani'
	    when price >= (select max(price)/3 from cars) and price < (select 2*max(price)/3 from cars) then 'Średni'
	    else 'Drogi' end
from cars
```
